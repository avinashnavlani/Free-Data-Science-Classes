# Stock Price Streaming

```python
import random
import time
from random import choice

class StockPriceStream:
  def __init__(self, stocks, max_updates=10):
    self.stocks = stocks
    self.max_updates = max_updates
    self.count = 1

  # def __iter__(self):
  #   return self

  def __next__(self):
    if self.count <= self.max_updates:
      stock = random.choice(self.stocks)
      price = round(random.uniform(100,500),2)
      timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
      self.count +=1
      # print(f"{timestamp} | {stock} - {price}")
      return f"{timestamp} | {stock} - {price}"
    else:
      raise StopIteration
```

```python
stocks = ["ASIAN PAINTS", "BHARTI AIRTEL", "NESTLE", "SBI", "AXIS"]

stock_itr = StockPriceStream(stocks)
```

```python
stock_itr.__next__()
```

```python
next(stock_itr)
```

```python
next(stock_itr)
```

```python
next(stock_itr)
```

```python
for stock in stock_itr:
  print(stock)
  time.sleep(1)
```
