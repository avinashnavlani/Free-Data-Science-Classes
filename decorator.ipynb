```python
def decorator_demo(func):
  def wrapper(*args, **kwargs):
    # before functionality
    print("Before function call")

    output=func(*args, **kwargs)

    # after functionality
    print("After function call")

    return output

  return wrapper
```

```python
def cube(num):
  return num**3
```

```python
decorator = decorator_demo(cube)
```

```python
out = decorator(4)
```

```python
out
```

```python
@decorator_demo
def cube(num):
  return num**3
```

```python
cube(3)
```

```python

# calculate execution time
import time

def execution_time(func):
  def wrapper(*args, **kwargs):
    start_time = time.time()

    output = func(*args, **kwargs)
    print("Cube: ", output)

    end_time = time.time()

    print("Execution time:", (end_time - start_time))
  return wrapper
```

```python
@execution_time
def cube(num):
  return num**3
```

```python
cube(7)
```
