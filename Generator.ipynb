```python
import random
import time

def stock_price_stream(stocks, max_updates=10):
  count = 1
  while count < max_updates:
    stock = random.choice(stocks)
    price = round(random.uniform(100,500),2)
    timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
    count +=1
    yield f"{timestamp} | {stock} - {price}"
```

```python
stocks = ["Adani", "Reliance", "Wipro", "HDFC"]
```

```python
stock_price_gen = stock_price_stream(stocks)
```

```python
next(stock_price_gen)
```
